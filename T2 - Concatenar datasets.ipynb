```python
# Olympics medals dataset
medals = pd.read_csv('../datasets/athletes/Medals.csv', encoding='ISO-8859-1')

medals.head()
```

```python
countries = pd.read_csv('../datasets/athletes/Athelete_Country_Map.csv', encoding='ISO-8859-1')
countries.head()
```

```python
# Data Sports
sports = pd.read_csv('../datasets/athletes/Athelete_Sports_Map.csv', encoding='ISO-8859-1')
sports.head()
```

```python
# Merge de los datasets
medals_countries = pd.merge(left=medals, right=countries, on='Athlete')
medals_countries.head()
```

```python
medals_countries_sports = pd.merge(left=medals_countries, right=sports, on='Athlete')
medals_countries_sports.head()
```

```python
# En el inner join se duplican los datos de los atletas que ganaron medallas en más de un deporte
medals_countries_sports[medals_countries_sports.Athlete == 'Aleksandar Ciric']
```

```python
countries_dropped = countries.drop_duplicates(subset='Athlete')
medals_countries = pd.merge(left=medals, right=countries_dropped, on='Athlete')
medals_countries[medals_countries.Athlete == 'Aleksandar Ciric']
```

```python
# Drop sports
sports_dropped = sports.drop_duplicates(subset='Athlete')
medals_countries_sports = pd.merge(left=medals_countries, right=sports_dropped, on='Athlete')
len(medals) == len(medals_countries_sports) # No hay atletas duplicados
```

```python
medals_countries_sports.head()
```

```python
# Formas de cruzar datasets (joins)
# El Inner Join nos va devolver la intersección de los datasets (sólo los datos que coinciden)
# El left join nos va a devolver un dataframe con los datos del dataset de la izquierda y los datos que coincidan con el de la derecha
# El right join nos va a devolver un dataframe con los datos del dataset de la derecha y los datos que coincidan con el de la izquierda

# Escogemos 6 atletas al azar
out_athletes = np.random.choice(medals['Athlete'], size=6, replace=False)
data_country_dlt = countries[~countries['Athlete'].isin(out_athletes) & (countries['Athlete'] != 'Michael Phelps')]
data_sports_dlt = sports[~sports['Athlete'].isin(out_athletes) & (sports['Athlete'] != 'Michael Phelps')]
data_medals_dlt = medals[~medals['Athlete'].isin(out_athletes) & (medals['Athlete'] != 'Michael Phelps')]

# Inner join
medals_inner = pd.merge(left=medals, right=data_country_dlt, how='inner', on='Athlete')

len(medals_inner)
```

```python
medals_inner.head()
```

```python
# Left join
medals_left = pd.merge(left=medals, right=data_country_dlt, how='left', on='Athlete')
medals_left.head()
```

```python
# Right join
medals_right = pd.merge(left=data_medals_dlt, right=countries, how='right', on='Athlete')
medals_right.head()
```

```python
data_country_yo = data_country_dlt.append({'Athlete': 'YO', 'Country': 'Shile'}, ignore_index=True)

# Outer join
medals_outer = pd.merge(left=data_medals_dlt, right=data_country_yo, how='outer', on='Athlete')
medals_outer.tail(10)
```
