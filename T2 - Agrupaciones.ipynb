```python
# Para crear un dataframe con datos aleatorios 
# Se pueden usar diversas funciones aleatorias
import pandas as pd
import numpy as np

df_len = 1000
age = np.random.randint(18, 80, df_len)
economic_status = np.random.choice(['Middle Class', 'Poor', 'Rich'], df_len)
height = np.random.normal(1.75, 0.20, df_len)
gender = np.random.choice(["Male", "Female"], df_len)
income = np.random.normal(27000, 15000, df_len)
weight = np.random.normal(70.3, 15, df_len)
# Se crea un diccionario con los datos
data = {
    'Age': age,
    'Economic Status': economic_status,
    'Gender': gender,
    'Height': height,
    'Income': income,
    'Weight': weight
}
# Creamos un dummy df
df = pd.DataFrame(data)

df.head()
```

```python
# Agrupacion de datos
# Agrupar por genero
group_gender = df.groupby('Gender')
# La funcion Groupby no divide el dataframe en grupos
# Solo crea un objeto de agrupacion

for names, groups in group_gender:
    print(names)
    print(groups)
```

```python
double_group = df.groupby(["Gender", "Economic Status"])
len(double_group)
```

```python
# Agregación de datos, Agregar es aplicar una funcion a cada grupo y obtener un resultado para cada grupo
double_group.sum()
```

```python
double_group.mean()
```

```python
double_group.size()
```

```python
double_group.describe()
```

```python
# Podemos seleccionar una de las columnas para obtener un resultado

grouped_income = double_group["Income"]

grouped_income.describe()
```

```python
# El metodo aggregate nos permite aplicar varias funciones a los datos
# Se puede usar una lista de funciones o un diccionario

double_group.aggregate(
    {
        "Income": np.sum,
        "Age": np.mean,
        "Height": np.std
    }
)
```

```python
# Tipificacion de la edad
double_group.aggregate(
    {
        "Age": np.mean,
        "Height": lambda h: np.mean(h) / np.std(h)
    }
)
```

```python
double_group.aggregate([np.sum, np.mean, np.std])
```

```python
double_group['Age'].filter(lambda x: x.sum() > 2400)
```

```python
zscore = lambda x: (x - x.mean()) / x.std()
z_group = double_group.transform(zscore)
```

```python
z_group.head()
```

```python
import matplotlib.pyplot as plt
plt.hist(z_group['Age'], bins=100)
```

```python
fill_na_mean = lambda x: x.fillna(x.mean())
double_group.transform(fill_na_mean)
```
