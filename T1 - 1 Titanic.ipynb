```python
from collections import defaultdict
class FileIter():
    def __init__(self, path, chunksize=20):
        self.path = path
        self.f = open(path)
        self.chunksize = chunksize
        self.headers = self.f.readline().strip().split(',')

    def __del__(self):
        self.f.close()
    
    def __iter__(self):
        return self
    
    def __next__(self):
        data = defaultdict(list)
        for i in range(self.chunksize):
            line = self.f.readline().strip().split(',')
            if not line:
                break
            for i, h in enumerate(self.headers):
                data[h].append(line[i])
        if not data:
            f.close()
            raise StopIteration
        return pd.DataFrame(data)
```
